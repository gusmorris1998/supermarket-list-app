### Proxy Main Configurations

Runs the proxy port in the terminal


__Port__: The port on which the proxy manager itself will listen for incoming proxy requests. For example if you're running the proxy manager locally and want it to handle requests on port 22999, you would use ```--port 22999```  
__Sticky_IP__: Ensures that the same IP address is used for requests from a given host  
__Proxy_Type__: Options can be used to persist proxy setting across sessions or configurations  
__Proxy_Port__: The port for the super proxy that the Proxy Manager connects to. The super proxy is the backend proxy server ```"brd.superproxy.io"``` that the Proxy Manager uses to route traffic through


````python
import subprocess

port = 24000
proxy_port = 24001
sticky_ip = "false"
proxy_type = "persist"




# Start the server in the background
process = subprocess.Popen(["proxy-manager", 
                            "-p", str(port),
                            "sticky_ip", sticky_ip,
                            "proxy_type", proxy_type,
                            ])

# Save the process ID
pid = process.pid
print(f"Server started with PID: {pid}")
````

### Killing the Server

To kill the server, use `process.terminate()` or `os.kill` with the saved PID:

````python
def kills(pid):
    '''Kills all processes'''
    parent = psutil.Process(pid)
    for child in parent.children(recursive=True):
        child.kill()
    parent.kill()

kills(pid)
````

### Pydantic

By using Pydantic with our webscraper we can create __Schema__'s for the data that we draw from API's and we can then perform serialization and validation.

__Schema:__ A schema is a blueprint or structure for our data that will define how it is organized within a database  
> __Serialization:__ is the process of converting data into a specific format (like JSON or XML) based on a schema, making it ready for storage, transmission, or further processing.
> __Validation:__ is the process of checking data against a schemas rules (such as data types, required fields, or constraints) to ensure it is correct, consistent, and meets predefined standards before storage or processing.
